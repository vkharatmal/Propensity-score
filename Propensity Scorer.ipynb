```python
dr_droped_df=droped_df.drop(columns=['R_rank_norm','F_rank_norm','M2_rank_norm'])
pd.Series([vif(dr_droped_df.values, i) 
               for i in range(dr_droped_df.shape[1])], 
              index=dr_droped_df.columns)
```

```python
import numpy as np
a=np.random.choice(dr_droped_df[dr_droped_df.Product_Activation==0].index,size=len(dr_droped_df[dr_droped_df.Product_Activation==1]),replace=False)
```

```python
_=pd.DataFrame(dr_droped_df,index=a)
```

```python
_.shape
```

```python
temp=dr_droped_df[dr_droped_df.Product_Activation==1]
```

```python
final_df=pd.concat([_,temp], axis=0, ignore_index=True)
```

```python
final_df
```

```python
final_df.Product_Activation.value_counts()
```

```python
final_df
```

```python
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.model_selection import train_test_split
```

```python
xTrain,xTest,yTrain,yTest=train_test_split(final_df.drop(columns='Product_Activation',axis=1),final_df.Product_Activation,random_state=42)
```

```python
model=RF(n_estimators=10,criterion='entropy',max_depth=12)
model.fit(xTrain,yTrain)
```

```python
model=RF(n_estimators=10,criterion='gini',max_depth=12)
model.fit(xTrain,yTrain)
```

```python
model.score(xTest,yTest)
```

```python
results=model.predict(xTest)
```

```python
prob=model.predict_proba(xTest)
```
